```python
import csv
import json
```

```python
with open('/content/drive/MyDrive/sarcasm_nonsarcasm/Sarcasm_Headlines_Dataset.json', 'r') as f:
    # Initialize an empty list to store the JSON objects
    objects = []
    # Loop through each line in the file
    for line in f:
        # Load the JSON object from the line
        obj = json.loads(line)
        # Append the object to the list
        objects.append(obj)

# Now you have a list of JSON objects that you can work with
print(objects)
```

```python
csv_file = '/content/drive/MyDrive/sarcasm_nonsarcasm/sarcasm_nonsarcasm.csv'
```

```python
with open(csv_file, 'w', newline='', encoding='utf-8') as f:
    # Create a CSV writer object
    writer = csv.writer(f)

    # Write the header row
    writer.writerow(['is_sarcastic', 'headline', 'article_link'])

    # Write each row of data to the CSV file
    for row in objects:
        writer.writerow([row['is_sarcastic'], row['headline'].encode('utf-8').decode('unicode_escape'), row['article_link']])

print(f'Successfully converted JSON file to CSV: {csv_file}.')
```
